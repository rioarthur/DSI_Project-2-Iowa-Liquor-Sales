**Model 1: Analyzing To See The Impact County Has On Liquor Sales (Using Ordinary Least Squares)**

```python
import statsmodels.api as sm
```

```python
dep = liquor_s['Sale (Dollars)'] #Create the dependent and independent variables for our first model
indep = sm.add_constant(liquor_dum) 
```

```python
indep.head()
```

```python
model = sm.OLS(dep,indep) #Run the variables in our OLS model
results = model.fit()
results.summary()
```

This model shows that using county alone does not do a good job of fitting the sales data. The R-squared tells us very little of the variablity in sales (0.6%). However, we can see that based upon the model's co-efficients for some counties, Dallas has the highest one (148).

Mean Squared Error (MSE) Testing For Model 3


We want to see what the residuals are.

```python
#In the next steps, we're going to prepare the variables for MSE testing

import sklearn.metrics as metrics
import sklearn.linear_model as linear_model
import pandas as pd
lm = linear_model.LinearRegression(fit_intercept=True,
                                   normalize=False,
                                   copy_X=True,
                                   n_jobs=1)
```

```python
lm.fit(X=indep_3, y=dep)
```

```python
lm.coef_
```

```python
lm.intercept_
```

```python
predicted = lm.predict(X=indep_3)
```

```python
true = dep
```

```python
metrics.mean_squared_error(y_pred = predicted,
                           y_true = true)
```

This model is all relative to Adair County because we excluded one dummy variable from our regression (to prevent falling into the dummy variable trap). The county that has the highest beta coefficient is Dallas County. Based on this model, we would say that opening a liquor store in Dallas County could potentially lead to more sales than opening a liquor store in other counties. Also, we would say that for each one unit increase in bottles sold and sale retail price, sales would increase by 14.29 and 2.34 dollars respectively. We will need to minimize the MSE.

**Training/Testing Data For Model 3 (Only Model We Will Consider)**

```python
#import necessary train/test models
from sklearn import datasets
from sklearn import linear_model
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score, cross_val_predict
from sklearn import metrics
```

```python
X_train, X_test, y_train, y_test = train_test_split(indep_3, dep, test_size=0.3)
print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

```python
lm = linear_model.LinearRegression()
model_sk = lm.fit(X_train, y_train)
predictions = lm.predict(X_test)
```

K-Folds Cross Validation Method

```python
scores = cross_val_score(model_sk, X_train, y_train, cv=5) #to perform 5 tests
print("CV scores:", scores)
print("Average: ", scores.mean())
```

```python
indep_3.head()
```

```python
#method to predict on data
predictions = cross_val_predict(model_sk, indep_3, dep, cv=5)
plt.scatter(dep, predictions)
plt.xlabel('Actual')
plt.ylabel('Predicted')
accuracy = metrics.r2_score(dep, predictions)
print("Cross-Predicted Accuracy:", accuracy)
```

Regularization 

To aid in reducing the MSE.

```python
#Using Lasso regression (to reduce our MSE)
from sklearn.linear_model import LassoCV
from sklearn.linear_model import Lasso
lassoregcv = LassoCV(n_alphas=10, normalize=True, random_state=9)
lassoregcv.fit(X_train, y_train)
lassoregcv.alpha_
```

```python
lasso_reg = Lasso(alpha=0.001, normalize=True)
lasso_reg.fit(X_train, y_train)
lasso_ = lasso_reg.coef_
lasso_
```

```python
labeled = zip(indep_3.columns, lasso_)
labeled
```

```python
y_pred = lasso_reg.predict(X_test)
print(np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```

Using a lasso regression, The MSE has dropped significantly. It was originally 45466.3 with OLS estimates. With regularization, it is now close to 200.

Our recommendation would be to open a store in Dallas County and maximize the number of bottles sold. Number of bottles sold is a better predictor of high sales than other indicators.
